```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
```

```python
data = pd.DataFrame(pd.read_csv('../day5/data/svm_hyperplane.csv'))
data.head()
```

```python
data_y = data[data['sold'] == 'Yes']
x1 = data_y['size']
x2 = data_y['price']
plt.scatter(x1,x2,color='r',label='Yes')

data_n = data[data['sold'] == 'No']
x1 = data_n['size']
x2 = data_n['price']
plt.scatter(x1,x2,color='w',label='No', marker='o', edgecolors='r')
plt.legend(loc='best')

x = data[['size', 'price']]
y = data['sold']
model = SVC(kernel='linear')
model.fit(x,y)

b = model.intercept_[0]
w1 = model.coef_[0,0]
w2 = model.coef_[0,1]

# Plot Hyperplane : b + w1x1 + w2x2 = 0
x1 = data['size']
x2 = (-b -w1 * x1) / w2
plt.plot(x1,x2,'g-')

# Plot Positive Boundary : b + w1x1 + w2x2 = 1
x2 = (1 -b -w1 * x1) / w2
plt.plot(x1,x2,'b--')

# Plot Negative Boundary : b + w1x1 + w2x2 = 1
x2 = (-1 -b -w1 * x1) / w2
plt.plot(x1,x2,'r--')
```
