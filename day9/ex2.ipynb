```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/SUT_AI_training/day9/ex2_icecream_sales_data.csv')
data.head()
```

```python
data.info()
```

```python
X = data['Temperature']
y = data['Revenue']
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(2, input_shape = [1]))
model.add(tf.keras.layers.Dense(1,))
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(0.5), loss='mean_squared_error')
```

```python
model.fit(X, y, epochs=50, verbose='none')
```

```python
sns.scatterplot(data = data, x = "Temperature", y = "Revenue")
```

```python
test = np.array([13.4,13,41,43,5])
y_pred= model.predict(test)
print(y_pred)
```

```python
plt.scatter(X,y)
plt.plot(test, y_pred, color='orange')
```
