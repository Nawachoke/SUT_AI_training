```python
cap = cv2.VideoCapture('../day7//data/capture.mp4')
while (cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture('../day7/data/capture.mp4')
while (cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        frame = cv2.resize(frame, (400, 400))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
fourcc  = cv2.VideoWriter_fourcc(*'MJPG')
rec = cv2.VideoWriter('../day7/data/capture.mp4', fourcc, 30.0, (640, 480))
while (cap.isOpened()):
    check, frame = cap.read()
    if check:
        frame = cv2.flip(frame, 1)
        rec.write(frame)
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
rec.release()
cap.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('../day7/data/cat.jpg')
img = cv2.resize(img, (700, 700))
cv2.line(img, (150, 180), (400, 430), (255, 0, 0), 5)
cv2.arrowedLine(img, (20, 0), (500, 500), (0, 255, 0), 2)
cv2.arrowedLine(img, (0, 0), (700, 700), (0, 0, 255), 1)
cv2.putText(img, 'cat', (10, 500), cv2.FONT_HERSHEY_SIMPLEX, 4, (0,0,0), 10)
cv2.rectangle(img, (100, 100), (500, 205), (0, 100, 100), 3)
cv2.circle(img, (300, 300), 100, (0, 255, 255), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import datetime
cap = cv2.VideoCapture('../day7/data/shibuya.mp4')
while (cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        frame = cv2.resize(frame, (400, 400))
        cv2.putText(frame, 'shibuya', (40, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,0,0), 10)
        dt = datetime.datetime.now()
        cv2.putText(frame, dt.strftime('%Y-%m-%d %H:%M:%S'), (40, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 2)
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
def click_event(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, f'{x}, {y}', (x, y), font, 1, (0, 0, 255), 2)
        cv2.imshow('image', img)
img = cv2.imread('../day7/data/cat.jpg')
cv2.imshow('image', img)
cv2.setMouseCallback('image', click_event)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
